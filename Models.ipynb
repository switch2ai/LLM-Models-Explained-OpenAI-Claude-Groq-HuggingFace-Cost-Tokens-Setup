```python
## How to calculate required space?

# 0,1

1 bit --> 1 transistor (1/0)
```

```python
A --> 65

10      --> 2 bits
11      --> 2 bits
111     --> 3 bits
1000001  --> 7 bits

## 8 bit = 1 byte

kb  --> 10**3 bytes
Mb  --> 10**6 bytes
Gb  --> 10**9 bytes
Tb  --> 10**12 bytes
--------------------------------------

## Model :

float32  --> 32 bits  = 4 byte
float64  --> 64 bits  = 8 byte
int8     --> 8 bits   = 1 byte

To store 1 model parameter  --> it uses 32 bits --> 4 bytes
To store 8 billion parameters --> 32 billion bytes --> 32 GB
```
